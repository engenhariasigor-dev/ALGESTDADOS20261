```python
#Atv-1 Peça ao usuário que digite sua idade usando input(), converta o valor para inteiro utilizando int() e exiba na tela a idade somada com 5 anos.
idade= int(input('Digite sua idade: '))
print(f"Sua idade daqui a cinco anos será: {idade+5} anos")
```

```python
#Atv-2 Solicite ao usuário sua altura em metros, converta o valor digitado para float e mostre o dobro da altura informada.
H= float(input('Digite sua altura em metros: '))
print (f"O dobro da sua altura é: {H*2:.2f} metros")
```

```python
#Atv-3 Crie uma variável inteira com valor 100, converta esse valor para str e concatene com a frase " pontos alcançados", exibindo o resultado final.
x= str(100)
print(x + " pontos encontrados")
```

```python
#Atv-4 Peça ao usuário que digite 1 para verdadeiro ou 0 para falso, converta primeiro para inteiro e depois para bool, mostrando o valor booleano resultante.
resultado= bool(int(input("Digite 1 para verdadeiro ou 0 para falso:")))
print(f"O valor booleano é {resultado}")
```

```python
#Atv-5 Peça três números ao usuário, converta todos para inteiro e mostre o maior entre eles.
pri= int(input('1º: '))
seg= int(input('2º: '))
ter= int(input('3º: '))
print(f"O maior dos três números é: {max(pri,seg,ter)}")
```

```python
#Atv-6 Peça quatro números decimais ao usuário, converta para float e calcule a média aritmética entre eles.
n1= float(input('1º Número Decimal: '))
n2= float(input('2º Número Decimal: '))
n3= float(input('3º Número Decimal: '))
n4= float(input('4º Número Decimal: '))
print(f"A média destes números decimais é: {((n1+n2+n3+n4)/4):.2f}")
```

```python
#Atv-7 Peça a idade do usuário em anos, converta para inteiro e calcule aproximadamente quantos meses essa pessoa já viveu.
idade= int(input('Idade: '))
print(f"Você já viveu {idade*12} meses")
```

```python
#Atv-8 Peça ao usuário que digite um número inteiro, converta esse número para str e utilize a função len() para mostrar quantos caracteres foram digitados.
nI= str(int(input('Número Inteiro: ')))
print(f'O número citado tem {len(nI)} caracteres')
```

```python
#Atv-10 Peça um número decimal ao usuário, converta para float e depois converta esse valor para inteiro usando int(), exibindo ambos os resultados para comparar a diferença.
nD= input('Número Decimal: ')
nD1= float(nD)
nD2= int(nD)
nD1,nD2
```

```python
#Atv-11 Solicite um número ao usuário, converta para inteiro e depois para bool, explicando em quais situações o resultado será False.
nU= int(input('Número: '))
boolean= bool(nU)
print(f'{boolean}, pois apenas valores 0 ou vazio é False em Python')
```

```python
#Atv-12 Peça ao usuário uma temperatura em Kelvin, converta para float, transforme o valor para Celsius e exiba o resultado.
tK= float(input('Informe uma temperatura em K°: '))
print(f"Em C° a temperatura é: {tK - 273.15:.2f}°")
```

```python
#Atv-13 Desenvolva um pequeno programa que receba nome, idade e salário do usuário, convertendo idade para inteiro e salário para float, e mostre todas as informações organizadas na tela.
nome= input('Nome: ')
idade= int(input('Idade: '))
sal= float(input('Salário: R$ '))
print(f'\n---FICHA CADASTRAL---\n\nNome: {nome}\nIdade: {idade}\nSalário: R$ {sal:.2f}')
```

```python
#Atv-14 Solicite o preço de um produto e o percentual de desconto, converta ambos para float, calcule o valor final com desconto aplicado e exiba o resultado.
pB= float(input('Valor Total: R$ '))
D= float(input('Desconto Percentual(%): '))
pD= pB - (pB*D)/100
print(f'O preço descontado é: {pD}')
```
