```python
# 01 - Classificador de Três Números
# Leia três números inteiros. Sem usar listas, determine e exiba qual é o maior, qual é o menor e qual é o do meio. Use apenas variáveis, operadores relacionais e condicionais aninhadas. Considere também o caso em que dois ou mais números sejam iguais.

print('INSIRA TRÊS NÚMEROS INTEIROS:') # Linha de Texto para instrução.
n1 = int(input('Primeiro número: ')) # Inserção do primeiro número inteiro.
n2 = int(input('Segundo número: ')) # Inserção do segundo número inteiro.
n3 = int(input('Terceiro número: ')) # Inserção do terceiro número inteiro

# Cálculo de maior, menor e meio.
if n1 == n2 or n1 == n3 or n2 == n3:
  print('Tem números iguais.')
else:
  nmax = max(n1,n2,n3) # Seleção do número de maior valor dentre os digitados.
  nmin = min(n1,n2,n3) # Seleção do número de menor valor dentre os digitados.
  nmid = ((n1 + n2 + n3) - nmax - nmin) # Cálculo para encontrar a diferença entre a soma dos três algarismos inseridos e seus respectivos extremos.
  print(f'Dentre os três números inseridos:\n{nmax} é o maior;\n{nmid} é o do meio;\n{nmin} é o menor.') # Texto da resolução
```

```python
#09 - Conversor de Temperatura Completo
# Leia uma temperatura e a unidade de origem usando match/case ("C" para Celsius, "F" para Fahrenheit ou "K" para Kelvin). Valide que temperaturas em Kelvin não sejam negativas. Calcule e exiba a temperatura convertida nas outras duas escalas. Use variáveis descritivas como temp_celsius, temp_fahrenheit e temp_kelvin. Use case _ para unidade inválida.

temp = int(input('Insira uma Temperatura: '))
unid = str(input('Insira a Unidade de origem (C,F ou K): '))

match unid:
  case "C"|"c":
    print(f'Temperatura: {temp} C.')
    print(f'Conversão em Fahrenheit: {(temp*9/5)+32} F.')
    print(f'Conversão em Kelvin: {temp+273.15} K.')
  case "F"|"f":
    print(f'Temperatura: {temp} F.')
    print(f'Conversão em Celsius: {(temp-32)*5/9} C.')
    print(f'Conversão em kelvin: {(temp-32)*5/9+273.15} K.')
  case "K"|"k":
    if temp>=0:
      print(f'Temperatura: {temp} K.')
      print(f'Conversão em Celsius: {temp-273.15} C.')
      print(f'Conversão em Fahrenheit: {(temp-273.15)*9/5+32} K.')
    else:
      print('Para Kelvin, valores negativos são inválidos.')
  case _:
    print('unidade inválida.')
```

```python
#10 Classificador de IMC com Risco.
#Leia peso (kg) e altura (m). Valide ambos (peso > 0, 0.5 <= altura <= 2.5). Calcule o IMC. Com elif, classifique em Abaixo do peso, Normal, Sobrepeso ou Obesidade (Grau I, II, III). Com if aninhado em cada classificação, informe o risco cardiovascular: Abaixo/Normal → baixo, Sobrepeso/Obesidade I → moderado, Obesidade II/III → alto. Exiba o IMC com 2 casas decimais.
h = 0
while not (0.5 <= h <= 2.5):
    h = float(input('Digite sua altura em metros (ex: 1.75): '))
    if not (0.5 <= h <= 2.5):
        print('Altura inválida. Por favor, digite um valor entre 0.5 e 2.5 metros.')

m = 0
while not (m > 0):
    m = float(input('Digite seu peso em quilogramas (ex: 70.5): '))
    if not (m > 0):
        print('Peso inválido. Por favor, digite um valor maior que zero.')

imc = m / (h ** 2)
classificacao = ""
risco = ""
print(f"\nSeu IMC é: {imc:.2f}")

if imc < 18.5:
    classificacao = "Abaixo do peso"
    risco = "baixo"
elif 18.5 <= imc < 24.9:
    classificacao = "Normal"
    risco = "baixo"
elif 25 <= imc < 29.9:
    classificacao = "Sobrepeso"
    risco = "moderado"
elif 30 <= imc < 34.9:
    classificacao = "Obesidade (Grau I)"
    risco = "moderado"
elif 35 <= imc < 39.9:
    classificacao = "Obesidade (Grau II)"
    risco = "alto"
else:
    classificacao = "Obesidade (Grau III)"
    risco = "alto"

print(f"Classificação: {classificacao}")
print(f"Risco Cardiovascular: {risco}")
```

```python
#12 Calculadora de Desconto Progressivo.
#Leia o preço original de um produto e um código de cupom usando match/case ("PROMO10", "PROMO20", "PROMO30" ou "SEMCUPOM"). Com if aninhado, verifique se o preço já é menor que R$50,00 — se sim, o desconto máximo permitido é 10%, independente do cupom. Exiba o cupom usado, o desconto em reais e o preço final.
preco_orig_str = input('Digite o preço original do produto: ')
cupom = input('Digite o código do cupom (PROMO10, PROMO20, PROMO30, CUPOM0): ')

preco_orig = float(preco_orig_str)
desc_perc = 0.0
cupom_ap = cupom.upper()

if preco_orig <= 0:
    print('Erro: O preço original deve ser um valor positivo.')
else:
    match cupom_ap:
        case 'PROMO10':
            desc_perc = 0.10
        case 'PROMO20':
            desc_perc = 0.20
        case 'PROMO30':
            desc_perc = 0.30
        case 'CUPOM0':
            desc_perc = 0.00
        case _:
            print(f'Cupom "{cupom}" inválido ou não reconhecido. Nenhum desconto aplicado.')
            cupom_ap = 'Nenhum'


    if preco_orig < 50.00:
        if desc_perc > 0.10:
            print('Atenção: Preço original menor que R$50,00. Desconto máximo permitido é de 10%.')
            desc_perc = 0.10

    desc_reais = preco_orig * desc_perc
    preco_fin = preco_orig - desc_reais

    print(f'\nCupom utilizado: {cupom_ap}')
    print(f'Desconto em reais: R$ {desc_reais:.2f}')
    print(f'Preço final: R$ {preco_fin:.2f}')
```

```python
#13 - Avaliador de Triângulo Completo
# Leia os três lados. Valide que todos são positivos na camada externa. Na camada interna, verifique a desigualdade triangular (cada lado deve ser menor que a soma dos outros dois). Se válido, use elif para classificar: equilátero (3 lados iguais), isósceles (exatamente 2 iguais) ou escaleno (todos diferentes). Se inválido, exiba mensagem clara.

lado1 = float(input('Insira o primeiro lado do triângulo: '))
lado2 = float(input('Insira o segundo lado do triângulo: '))
lado3 = float(input('Insira o terceiro lado do triângulo: '))

if (lado1 > 0) and (lado2 > 0) and (lado3 > 0):
    if (lado1 < lado2 + lado3) and (lado2 < lado1 + lado3) and (lado3 < lado1 + lado2):

        if (lado1 == lado2) and (lado2 == lado3):
            print('Triângulo equilátero.')
        elif (lado1 == lado2) or (lado1 == lado3) or (lado2 == lado3):
            print('Triângulo isósceles.')
        else:
            print('Triângulo escaleno.')

    else:
        print('Os valores não formam um triângulo.')

else:
    print('Todos os lados devem ser positivos.')
```

```python
#14 Calculadora de Notas com Conceito.
#Leia as quatro notas bimestrais de um aluno (0 a 10 cada). Valide todas. Calcule a média aritmética. Use elif para o conceito: média ≥ 9 → A, ≥ 7 → B, ≥ 5 → C, ≥ 3 → D, abaixo → E. Com if aninhado, exiba o status: A e B → "Aprovado", C → "Recuperação", D e E → "Reprovado". Exiba a média com 2 casas decimais e o conceito.
soma_notas = 0
for i in range(1, 5):
    nota = -1
    while not (0 <= nota <= 10):
        nota = float(input(f'Digite a nota {i} (0 a 10): '))
        if not (0 <= nota <= 10):
            print('Nota inválida. Digite um valor entre 0 e 10.')
    soma_notas = soma_notas + nota

media = soma_notas / 4
conceito = ''
if media >= 9:
    conceito = 'A'
elif media >= 7:
    conceito = 'B'
elif media >= 5:
    conceito = 'C'
elif media >= 3:
    conceito = 'D'
else:
    conceito = 'E'

status = ''
if conceito == 'A' or conceito == 'B':
    status = 'Aprovado'
elif conceito == 'C':
    status = 'Recuperação'
else:
    status = 'Reprovado'

print(f'\nMédia: {media:.2f}')
print(f'Conceito: {conceito}')
print(f'Status: {status}')
```

```python
#15 - Verificador de Acesso por Horário
# Leia a hora atual (0-23) e o tipo de usuário usando match/case ("admin", "operador", "visitante"). Valide a hora. Para cada tipo, use if aninhado com elif para verificar o turno (manhã: 6-11, tarde: 12-17, noite: 18-23, madrugada: 0-5). Regras: admin acessa sempre, operador apenas manhã e tarde, visitante apenas tarde. Exiba se o acesso é permitido ou negado e o turno atual.

hora = int(input('Insira a hora atual (0-23): '))
tipo = input('Insira o tipo de usuário (admin, operador, visitante): ')

if (hora >= 0) and (hora <= 23):

    match tipo:

        case "admin":
            if (hora >= 6) and (hora <= 11):
              turno = "manhã"
            elif (hora >= 12) and (hora <= 17):
              turno = "tarde"
            elif (hora >= 18) and (hora <= 23):
              turno = "noite"
            else:
              turno = "madrugada"

            print(f'Acesso permitido (admin). Turno atual: {turno}.')

        case "operador":
            if (hora >= 6) and (hora <= 11):
              turno = "manhã"
              print(f'Acesso permitido (operador). Turno atual: {turno}.')
            elif (hora >= 12) and (hora <= 17):
              turno = "tarde"
              print(f'Acesso permitido (operador). Turno atual: {turno}.')
            elif (hora >= 18) and (hora <= 23):
              turno = "noite"
              print(f'Acesso negado (operador). Turno atual: {turno}.')
            else:
              turno = "madrugada"
              print(f'Acesso negado (operador). Turno atual: {turno}.')

        case "visitante":
            if (hora >= 6) and (hora <= 11):
              turno = "manhã"
              print(f'Acesso negado (visitante). Turno atual: {turno}.')
            elif (hora >= 12) and (hora <= 17):
              turno = "tarde"
              print(f'Acesso permitido (visitante). Turno atual: {turno}.')
            elif (hora >= 18) and (hora <= 23):
              turno = "noite"
              print(f'Acesso negado (visitante). Turno atual: {turno}.')
            else:
              turno = "madrugada"
              print(f'Acesso negado (visitante). Turno atual: {turno}.')

        case _:
            print('Tipo de usuário inválido.')

else:
    print('Hora inválida. Insira um valor entre 0 e 23.')
```

```python
#16 Simulador de Bomba de Combustível.
#Leia o tipo de combustível usando match/case ("gasolina", "etanol", "diesel") e a quantidade de litros. Valide que a quantidade é positiva. Cada combustível tem um preço por litro definido como variável no código. Com if aninhado, aplique 3% de desconto se o abastecimento for acima de 40 litros. Exiba o subtotal, o desconto (se houver) e o total a pagar.
PRECO_GAS = 5.80
PRECO_ETA = 3.90
PRECO_DIE = 6.20

tipo_comb = ""
while tipo_comb != "gasolina" and tipo_comb != "etanol" and tipo_comb != "diesel":
    tipo_comb = input("Digite o tipo de combustível (gasolina, etanol, diesel): ").lower()
    if tipo_comb != "gasolina" and tipo_comb != "etanol" and tipo_comb != "diesel":
        print("Tipo de combustível inválido. Por favor, escolha entre gasolina, etanol ou diesel.")

qtd_litros = 0.0
while qtd_litros <= 0:
    lit_str = input("Digite a quantidade de litros: ")
    qtd_litros = float(lit_str)

    if qtd_litros <= 0:
        print("A quantidade de litros deve ser positiva.")

preco_l = 0.0
match tipo_comb:
    case "gasolina":
        preco_l = PRECO_GAS
    case "etanol":
        preco_l = PRECO_ETA
    case "diesel":
        preco_l = PRECO_DIE

subtotal = qtd_litros * preco_l

desc = 0.0
if qtd_litros > 40:
    desc = subtotal * 0.03

total_a_pagar = subtotal - desc

print(f"\nTipo de Combustível: {tipo_comb.capitalize()}")
print(f"Quantidade de Litros: {qtd_litros:.2f}L")
print(f"Preço por Litro: R$ {preco_l:.2f}")
print(f"Subtotal: R$ {subtotal:.2f}")
if desc > 0:
    print(f"Desconto (3%): R$ {desc:.2f}")
else:
    print("Desconto: R$ 0.00")
print(f"Total a Pagar: R$ {total_a_pagar:.2f}")
```

```python
#17 - Classificador de Velocidade
# Leia a velocidade em km/h e o tipo de via usando match/case ("urbana" → limite 60, "rodovia" → limite 100, "autopista" → limite 120). Valide que a velocidade não é negativa. Com if aninhado com elif, verifique: dentro do limite, excesso leve (até 20% acima), excesso moderado (até 50% acima) ou excesso grave (acima de 50%). Exiba o limite da via, a velocidade e a situação.

velocidade = float(input('Digite a velocidade (km/h): '))
via = input('Digite o tipo de via (urbana, rodovia, autopista): ')

if (velocidade >= 0):

  match via:
    case "urbana":
      limite = 60
    case "rodovia":
      limite = 100
    case "autopista":
      limite = 120
    case _:
      print('Tipo de via inválido.')
      limite = None

  if (limite != None):

    print(f'Limite da via: {limite} km/h')
    print(f'Velocidade informada: {velocidade} km/h')

    if (velocidade <= limite):
      print(f'Situação: dentro do limite.')
    elif (velocidade <= limite * 1.2):
      print(f'Situação: excesso leve.')
    elif (velocidade <= limite * 1.5):
      print(f'Situação: excesso moderado.')
    else:
      print(f'Situação: excesso grave.')

else:
    print('Velocidade inválida. Insira um valor positivo.')
```

```python
#20 Simulador de Pedágio Inteligente
#Leia o tipo de veículo usando match/case ("moto", "carro", "caminhao", "onibus") e se possui tag de pagamento automático (s/n). Defina a tarifa base de cada veículo como variável. Com if aninhado, aplique 10% de desconto com tag. Leia também a hora atual, valide (0-23) e, com elif aninhado, aplique 20% de acréscimo em horário de pico (7-9 ou 17-19). Exiba a tarifa detalhada.
TAR_MOT = 5.00
TAR_CAR = 10.00
TAR_CAM = 15.00
TAR_ONI = 20.00

veiculo = ""
while (veiculo != "moto" and veiculo != "carro" and
       veiculo != "caminhao" and veiculo != "onibus"):
    veiculo = input("Digite o tipo de veículo (moto, carro, caminhao, onibus): ").lower()
    if (veiculo != "moto" and veiculo != "carro" and
        veiculo != "caminhao" and veiculo != "onibus"):
        print("Tipo de veículo inválido. Por favor, escolha entre moto, carro, caminhao ou onibus.")

tar_base = 0.0
match veiculo:
    case "moto":
        tar_base = TAR_MOT
    case "carro":
        tar_base = TAR_CAR
    case "caminhao":
        tar_base = TAR_CAM
    case "onibus":
        tar_base = TAR_ONI

tag = ""
while (tag != "s" and tag != "n"):
    tag = input("Possui tag de pagamento automático (s/n)? ").lower()
    if (tag != "s" and tag != "n"):
        print("Opção inválida. Por favor, digite 's' para sim ou 'n' para não.")

tar_atual = tar_base
desc = 0.0
if tag == "s":
    desc = tar_base * 0.10
    tar_atual = tar_base - desc

hora = -1
while not (0 <= hora <= 23):
    hora_s = input("Digite a hora atual (0-23): ")


    hora = int(hora_s)
    if not (0 <= hora <= 23):
        print("Hora inválida. Por favor, digite um valor entre 0 e 23.")

acrescimo = 0.0
if (7 <= hora <= 9) or (17 <= hora <= 19):
    acrescimo = tar_atual * 0.20
    tar_final = tar_atual + acrescimo
else:
    tar_final = tar_atual

print(f"\n--- Detalhes da Tarifa ---")
print(f"Tipo de veículo: {veiculo.capitalize()}")
print(f"Tarifa base: R$ {tar_base:.2f}")
if tag == "s":
    print(f"Desconto por tag (10%): R$ {desc:.2f}")
print(f"Tarifa após desconto: R$ {tar_atual:.2f}")
print(f"Hora atual: {hora:02d}:00")
if acrescimo > 0:
    print(f"Acréscimo horário de pico (20%): R$ {acrescimo:.2f}")
print(f"Tarifa final a pagar: R$ {tar_final:.2f}")
```

```python
#23 - Sistema de Pontuação de Jogo
# Leia a pontuação de três jogadores (0 a 1000 cada). Valide todas. Usando apenas variáveis e condicionais, determine: quem tem a maior pontuação (1º lugar), quem tem a menor (3º lugar) e quem é o do meio (2º lugar). Trate empates explicitamente. Calcule a diferença entre o 1º e o 3º. Exiba o ranking completo e a diferença.

p1 = int(input('Insira Pontuação do jogador 1: '))
p2 = int(input('Insira Pontuação do jogador 2: '))
p3 = int(input('Insira Pontuação do jogador 3: '))

if (p1 >= 0 and p1 <= 1000) and (p2 >= 0 and p2 <= 1000) and (p3 >= 0 and p3 <= 1000):

  # Maior valor
  if (p1 >= p2) and (p1 >= p3):
    primeiro = 'Jogador 1'
    maior = p1
  elif (p2 >= p1) and (p2 >= p3):
    primeiro = 'Jogador 2'
    maior = p2
  else:
    primeiro = 'Jogador 3'
    maior = p3

  # Menor valor
  if (p1 <= p2) and (p1 <= p3):
    terceiro = 'Jogador 1'
    menor = p1
  elif (p2 <= p1) and (p2 <= p3):
    terceiro = 'Jogador 2'
    menor = p2
  else:
    terceiro = 'Jogador 3'
    menor = p3

  # Meio (quem não é nem maior nem menor)
  if (primeiro != 'Jogador 1') and (terceiro != 'Jogador 1'):
    segundo = 'Jogador 1'
  elif (primeiro != 'Jogador 2') and (terceiro != 'Jogador 2'):
    segundo = 'Jogador 2'
  else:
    segundo = 'Jogador 3'

  diferenca = maior - menor

  print(f'\n1º lugar: {primeiro}')
  print(f'2º lugar: {segundo}')
  print(f'3º lugar: {terceiro}')
  print(f'Diferença: {diferenca}')

else:
  print('Valores inválidos.')
```

```python
#24 Calculadora de Reajuste Salarial.
#Leia o salário atual e o percentual de avaliação de desempenho (0 a 100). Valide ambos. Use elif para o reajuste: avaliação ≥ 90 → 15%, ≥ 75 → 10%, ≥ 60 → 5%, abaixo → 0%. Com if aninhado, verifique se o novo salário ultrapassa o teto de R$15.000 — se sim, ajuste ao teto e exiba aviso. Exiba salário anterior, percentual aplicado, reajuste em reais e novo salário.
sal = 0.0
while not (sal > 0):
    sal_str = input('Digite o salário atual: ')
    val_sal = True
    for char in sal_str:
        if not ('0' <= char <= '9' or char == '.'):
            val_sal = False
            break
    if val_sal:
        sal = float(sal_str)
        if not (sal > 0):
            print('Erro: O salário deve ser um valor positivo. Tente novamente.')
    else:
        print('Erro: Entrada inválida para salário. Por favor, digite um número.')

aval = -1.0
while not (0 <= aval <= 100):
    aval_str = input('Digite o percentual de avaliação de desempenho (0 a 100): ')
    val_aval = True
    for char in aval_str:
        if not ('0' <= char <= '9' or char == '.'):
            val_aval = False
            break
    if val_aval:
        aval = float(aval_str)
        if not (0 <= aval <= 100):
            print('Erro: A avaliação deve estar entre 0 e 100. Tente novamente.')
    else:
        print('Erro: Entrada inválida para avaliação. Por favor, digite um número.')

perc_reaj = 0.0
if aval >= 90:
    perc_reaj = 0.15
elif aval >= 75:
    perc_reaj = 0.10
elif aval >= 60:
    perc_reaj = 0.05
else:
    perc_reaj = 0.00

reaj_val = sal * perc_reaj
novo_sal = sal + reaj_val
teto_sal = 15000.00
if novo_sal > teto_sal:
    print('Aviso: O novo salário ultrapassou o teto de R$15.000,00 e foi ajustado.')
    novo_sal = teto_sal

print(f'\n--- Detalhes do Reajuste ---')
print(f'Salário anterior: R$ {sal:.2f}')
print(f'Percentual aplicado: {perc_reaj * 100:.0f}%')
print(f'Reajuste em reais: R$ {reaj_val:.2f}')
print(f'Novo salário: R$ {novo_sal:.2f}')
```

```python
#25 - Classificador de Triângulo com Ângulos
# Leia os três ângulos internos de um triângulo em graus. Valide que todos são positivos. Verifique se a soma é exatamente 180° — se não for, exiba "Triângulo inválido" e o valor da soma encontrada. Se válido, use elif para classificar quanto aos ângulos: acutângulo (todos < 90°), retângulo (algum == 90°) ou obtusângulo (algum > 90°). Com if aninhado, classifique também quanto aos lados com base nos ângulos iguais.

a1 = float(input('Ângulo 1: '))
a2 = float(input('Ângulo 2: '))
a3 = float(input('Ângulo 3: '))

if (a1 > 0) and (a2 > 0) and (a3 > 0):

  soma = a1 + a2 + a3

  if (soma == 180):

    # Classificação quanto aos ângulos
    if (a1 < 90) and (a2 < 90) and (a3 < 90):
      print('\nTriângulo acutângulo')

    elif (a1 == 90) or (a2 == 90) or (a3 == 90):
      print('\nTriângulo retângulo')

    else:
      print('\nTriângulo obtusângulo')

    # Classificação quanto aos lados (if aninhado)
    if (a1 == a2) and (a2 == a3):
      print('Equilátero')

    elif (a1 == a2) or (a1 == a3) or (a2 == a3):
      print('Isósceles')

    else:
      print('Escaleno')

  else:
    print('\nTriângulo inválido')
    print(f'Soma encontrada: {soma}')

else:
  print('Valores inválidos.')
```

```python
#26 Calculadora de Combustível de Viagem.
#Leia a distância em km, o consumo do veículo em km/l e o preço do combustível por litro. Valide todos como positivos. Calcule os litros necessários e o custo total. Com if aninhado, verifique se o custo ultrapassa R$500 — se sim, calcule e exiba o km sugerido para uma parada de reabastecimento na metade do percurso. Exiba todos os dados com 2 casas decimais.
def get_positive_float(prompt):
    val = 0.0
    while not (val > 0):
        str_input = input(prompt)
        is_valid = True
        dot_count = 0
        for char in str_input:
            if char == '.':
                dot_count = dot_count + 1
            elif not ('0' <= char <= '9'):
                is_valid = False
                break
        if dot_count > 1:
            is_valid = False

        if is_valid:
            val = float(str_input)
            if not (val > 0):
                print('Erro: O valor deve ser positivo. Tente novamente.')
        else:
            print('Erro: Entrada inválida. Por favor, digite um número.')
    return val

dist = get_positive_float('Digite a distância em km: ')
cons = get_positive_float('Digite o consumo do veículo em km/l: ')
prec = get_positive_float('Digite o preço do combustível por litro: ')
litros = dist / cons
custo_t = litros * prec

print(f'\n--- Detalhes da Viagem ---')
print(f'Distância: {dist:.2f} km')
print(f'Consumo: {cons:.2f} km/l')
print(f'Preço/Litro: R$ {prec:.2f}')
print(f'Litros necessários: {litros:.2f} L')
print(f'Custo total: R$ {custo_t:.2f}')

if custo_t > 500.00:
    km_parada = dist / 2
    print(f'Atenção: O custo total (${custo_t:.2f}) ultrapassa R$ 500,00.')
    print(f'Sugestão: Reabastecer na metade do percurso, aos {km_parada:.2f} km.')
```

```python
#27 - Verificador de Senha com Nível de Segurança
# Leia uma senha e uma confirmação. Na camada externa, verifique se as duas são iguais — se não, exiba "Senhas não conferem" e encerre a lógica. Se iguais, use elif para classificar: menos de 6 caracteres → "Fraca", 6-9 → "Média", 10 ou mais → "Forte". Com if aninhado, verifique se a senha é composta apenas por dígitos numéricos (use a condição senha.isdigit()) — se sim, rebaixe um nível. Exiba o nível final.

senha = input('Insira a senha: ')
confirmacao = input('Confirme a senha: ')

if (senha == confirmacao):

  tamanho = len(senha)

  # Classificação inicial
  if (tamanho < 6):
    nivel = 'Fraca'
  elif (tamanho <= 9):
    nivel = 'Média'
  else:
    nivel = 'Forte'

  # Verificação adicional (if aninhado)
  if (senha.isdigit()):

    if (nivel == 'Forte'):
      nivel = 'Média'
    elif (nivel == 'Média'):
      nivel = 'Fraca'

  print(f'\nNível da senha: {nivel}')

else:
  print('Senhas não conferem.')
```

```python
#28 Simulador de Reserva de Assento.
#Leia o número do assento (1 a 100) e a classe desejada usando match/case ("economica", "executiva", "primeira"). Valide o número. Com if aninhado, verifique se a faixa do assento corresponde à classe: 1-10 → primeira, 11-30 → executiva, 31-100 → econômica. Se incompatível, exiba mensagem. Se compatível, calcule o preço (econômica R$350, executiva R$890, primeira R$1.500) e exiba a confirmação.
PRIMEIRA_CLASSE_MIN_ASSENTO = 1
PRIMEIRA_CLASSE_MAX_ASSENTO = 10
EXECUTIVA_CLASSE_MIN_ASSENTO = 11
EXECUTIVA_CLASSE_MAX_ASSENTO = 30
ECONOMICA_CLASSE_MIN_ASSENTO = 31
ECONOMICA_CLASSE_MAX_ASSENTO = 100

PRECO_PRIMEIRA_CLASSE = 1500.00
PRECO_EXECUTIVA_CLASSE = 890.00
PRECO_ECONOMICA_CLASSE = 350.00
numero_assento = 0
while not (PRIMEIRA_CLASSE_MIN_ASSENTO <= numero_assento <= ECONOMICA_CLASSE_MAX_ASSENTO):
    entrada_assento_str = input("Digite o número do assento (1 a 100): ")
    e_digito = True
    for caractere in entrada_assento_str:
        if not ('0' <= caractere <= '9'):
            e_digito = False
            break
    if e_digito and entrada_assento_str != "":
        numero_assento_temp = int(entrada_assento_str)
        if (PRIMEIRA_CLASSE_MIN_ASSENTO <= numero_assento_temp and numero_assento_temp <= ECONOMICA_CLASSE_MAX_ASSENTO):
            numero_assento = numero_assento_temp
        else:
            print("Número de assento inválido. Deve ser entre 1 e 100.")
    else:
        print("Entrada inválida para o assento. Por favor, digite um número inteiro.")

classe_desejada = ""
while not (classe_desejada == "economica" or classe_desejada == "executiva" or classe_desejada == "primeira"):
    classe_desejada = input("Digite a classe desejada (economica, executiva, primeira): ").lower()
    if not (classe_desejada == "economica" or classe_desejada == "executiva" or classe_desejada == "primeira"):
        print("Classe inválida. Por favor, escolha entre 'economica', 'executiva' ou 'primeira'.")
preco_final = 0.0
mensagem_saida = ""
classe_real_do_assento = ""

match classe_desejada:
    case "primeira":
        if numero_assento >= PRIMEIRA_CLASSE_MIN_ASSENTO and numero_assento <= PRIMEIRA_CLASSE_MAX_ASSENTO:
            preco_final = PRECO_PRIMEIRA_CLASSE
            classe_real_do_assento = "Primeira"
            mensagem_saida = f"Reserva confirmada! Assento {numero_assento} na classe {classe_real_do_assento}. Preço: R$ {preco_final:.2f}"
        else:
            mensagem_saida = f"Assento {numero_assento} não corresponde à classe Primeira (faixa {PRIMEIRA_CLASSE_MIN_ASSENTO}-{PRIMEIRA_CLASSE_MAX_ASSENTO})."
    case "executiva":
        if numero_assento >= EXECUTIVA_CLASSE_MIN_ASSENTO and numero_assento <= EXECUTIVA_CLASSE_MAX_ASSENTO:
            preco_final = PRECO_EXECUTIVA_CLASSE
            classe_real_do_assento = "Executiva"
            mensagem_saida = f"Reserva confirmada! Assento {numero_assento} na classe {classe_real_do_assento}. Preço: R$ {preco_final:.2f}"
        else:
            mensagem_saida = f"Assento {numero_assento} não corresponde à classe Executiva (faixa {EXECUTIVA_CLASSE_MIN_ASSENTO}-{EXECUTIVA_CLASSE_MAX_ASSENTO})."
    case "economica":
        if numero_assento >= ECONOMICA_CLASSE_MIN_ASSENTO and numero_assento <= ECONOMICA_CLASSE_MAX_ASSENTO:
            preco_final = PRECO_ECONOMICA_CLASSE
            classe_real_do_assento = "Econômica"
            mensagem_saida = f"Reserva confirmada! Assento {numero_assento} na classe {classe_real_do_assento}. Preço: R$ {preco_final:.2f}"
        else:
            mensagem_saida = f"Assento {numero_assento} não corresponde à classe Econômica (faixa {ECONOMICA_CLASSE_MIN_ASSENTO}-{ECONOMICA_CLASSE_MAX_ASSENTO})."

print(mensagem_saida)
```

```python
#29 - Simulador de Batalha por Tipos
# Leia o tipo do atacante e o tipo do defensor usando match/case. Tipos: "fogo", "agua", "planta", "eletrico", "terra". Relações de vantagem: fogo vence planta, água vence fogo, planta vence água, elétrico vence água, terra vence elétrico. Com if aninhado, determine se é vantagem (dano x2), desvantagem (dano x0.5) ou neutro (dano x1). Leia o poder base do ataque, calcule o dano final e exiba o resultado.

print('SIMULADOR DE BATALHA POR TIPOS')
print('Tipos: fogo, agua, planta, eletrico e terra')
atacante = input('Tipo do Pokémon atacante (apenas um): ')
defensor = input('Tipo do Pokémon defensor (apenas um): ')

poder = float(input('Insira o Poder Base do ataque: '))

multiplicador = 1

match atacante:

  case 'fogo':
    if (defensor == 'planta'):
      multiplicador = 2
    elif (defensor == 'agua'):
      multiplicador = 0.5

  case 'agua':
    if (defensor == 'fogo'):
      multiplicador = 2
    elif (defensor == 'planta') or (defensor == 'eletrico'):
      multiplicador = 0.5

  case 'planta':
    if (defensor == 'agua'):
      multiplicador = 2
    elif (defensor == 'fogo'):
      multiplicador = 0.5

  case 'eletrico':
    if (defensor == 'agua'):
      multiplicador = 2
    elif (defensor == 'terra'):
      multiplicador = 0.5

  case 'terra':
    if (defensor == 'eletrico'):
      multiplicador = 2

  case _:
    print('Tipo inválido.')

dano = poder * multiplicador

print(f'Atacante: {atacante}     Defensor: {defensor}')
print(f'\nMultiplicador de dano por vantagem de tipagem: x{multiplicador}')
print(f'Dano final: {dano}')
```

```python
#30 Sistema Completo de Reserva de Hotel.
#Leia o tipo de acomodação usando match/case ("standard", "luxo", "suite"), o número de diárias (validar como positivo) e se o hóspede tem cartão fidelidade (s/n). Defina a diária base de cada tipo como variável. Use elif aninhado para desconto por diárias: 1-3 → 0%, 4-7 → 10%, acima de 7 → 20%. Com if aninhado adicional, aplique mais 5% para cartão fidelidade. Valide que o total não seja inferior a R$100. Use variáveis descritivas (diaria_base, desconto_qtd, desconto_fidelidade, total_final) e exiba o resumo completo com f-string.
PRECO_STANDARD = 150.00
PRECO_LUXO = 250.00
PRECO_SUITE = 400.00

tipo_acomodacao = ""
validacao_acomodacao_ok = False
while validacao_acomodacao_ok == False:
    tipo_acomodacao = input("Tipo de acomodação (standard, luxo, suite): ").lower()
    if (tipo_acomodacao == "standard" or tipo_acomodacao == "luxo" or tipo_acomodacao == "suite"):
        validacao_acomodacao_ok = True
    else:
        print("Tipo de acomodação inválido. Escolha 'standard', 'luxo' ou 'suite'.")

numero_diarias = 0
validacao_diarias_ok = False
while validacao_diarias_ok == False:
    entrada_diarias_str = input("Número de diárias: ")
    e_numero = True
    for caractere in entrada_diarias_str:
        if (caractere < '0' or caractere > '9'):
            e_numero = False
            break
    if e_numero and entrada_diarias_str != "":
        numero_diarias_temp = int(entrada_diarias_str)
        if numero_diarias_temp > 0:
            numero_diarias = numero_diarias_temp
            validacao_diarias_ok = True
        else:
            print("O número de diárias deve ser positivo.")
    else:
        print("Entrada inválida. Digite um número inteiro positivo.")

cartao_fidelidade = ""
validacao_fidelidade_ok = False
while validacao_fidelidade_ok == False:
    cartao_fidelidade = input("Tem cartão fidelidade (s/n)? ").lower()
    if (cartao_fidelidade == "s" or cartao_fidelidade == "n"):
        validacao_fidelidade_ok = True
    else:
        print("Entrada inválida. Digite 's' para sim ou 'n' para não.")

valor_diaria_base = 0.0
match tipo_acomodacao:
    case "standard":
        valor_diaria_base = PRECO_STANDARD
    case "luxo":
        valor_diaria_base = PRECO_LUXO
    case "suite":
        valor_diaria_base = PRECO_SUITE

subtotal_reserva = valor_diaria_base * numero_diarias
percentual_desconto_quantidade = 0.0
if numero_diarias >= 1 and numero_diarias <= 3:
    percentual_desconto_quantidade = 0.0
elif numero_diarias >= 4 and numero_diarias <= 7:
    percentual_desconto_quantidade = 0.10
else:
    percentual_desconto_quantidade = 0.20

valor_desconto_quantidade = subtotal_reserva * percentual_desconto_quantidade
total_apos_desconto_quantidade = subtotal_reserva - valor_desconto_quantidade
percentual_desconto_fidelidade = 0.0
if cartao_fidelidade == 's':
    percentual_desconto_fidelidade = 0.05

valor_desconto_fidelidade = total_apos_desconto_quantidade * percentual_desconto_fidelidade
valor_total_final = total_apos_desconto_quantidade - valor_desconto_fidelidade
if valor_total_final < 100.00:
    valor_total_final = 100.00
    print("Atenção: O valor total foi ajustado para o mínimo de R$100,00.")

print(f"\n--- Resumo da Reserva ---")
print(f"Tipo de Acomodação: {tipo_acomodacao.capitalize()}")
print(f"Número de Diárias: {numero_diarias}")
print(f"Diária Base: R$ {valor_diaria_base:.2f}")
print(f"Cartão Fidelidade: {'Sim' if cartao_fidelidade == 's' else 'Não'}")
print(f"Subtotal (sem descontos): R$ {subtotal_reserva:.2f}")
print(f"Desconto por Diárias ({percentual_desconto_quantidade*100:.0f}%): R$ {valor_desconto_quantidade:.2f}")
print(f"Desconto Fidelidade ({percentual_desconto_fidelidade*100:.0f}%): R$ {valor_desconto_fidelidade:.2f}")
print(f"Total Final a Pagar: R$ {valor_total_final:.2f}")
```
